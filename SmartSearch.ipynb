```python
from sentence_transformers import SentenceTransformer, util

# Load a pre-trained model, for example, 'all-MiniLM-L6-v2' or 'FinBERT'
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
# Sample documents
documents = [
    "Financial market trends and their impact on economic policy.",
    "Stock market analysis using machine learning techniques.",
    "Recent updates on federal interest rates.",
    "Understanding bonds and their role in investment portfolios."
]

# Embed the documents
document_embeddings = model.encode(documents, convert_to_tensor=True)
```

```python
# User query
query = "What are the recent trends in the stock market?"
query_embedding = model.encode(query, convert_to_tensor=True)

# Compute cosine similarity scores
cosine_scores = util.pytorch_cos_sim(query_embedding, document_embeddings)

# Rank documents based on similarity scores
import torch
ranked_results = torch.argsort(cosine_scores, descending=True)
for idx in ranked_results[0]:
    print(f"Document: {documents[idx]}, Score: {cosine_scores[0][idx]:.4f}")
```
